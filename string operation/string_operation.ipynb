## Operations on Strings

- Arithmetic Operations (concatenating)
- Relational Operations
- Logical Operations
- Loops on Strings
- Membership Operations

```python
# Arithmetic Operations

print("Hi" + " " + "Everyone") #concatination
```

```python
print("Hi" * 500)
```

```python
# Relational Operations

print("Hello" == "Hello") # equality
print("Hello"!= "World") # not equal
print("Hello" > "Hi") # greater than
print("Hello" < "World") # less than
print("Hello" >= "Hello") # greater than or equal to
```

```python
# Logical Operations

"hello" and "World"
```

```python
"hello" or 'world'
```

```python
not "hello"
```

```python
# Loops on Strings
for i in "Apurbo ":
    print(i)
```

```python
# Membership Operations

'p' in "Apurbo"
```

```python
# String formating

name = "Apurbo"
gender = "Male"

s = "My name is {}, and I am a {}"
```

```python
s.format(name, gender)
```

```python
# String formating

name = "Apurbo"
gender = "Male"

s = "My name is {0}, and I am a {1}"
```

```python
s.format(name, gender)
```

# F-String
