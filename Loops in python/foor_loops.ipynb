# print triangle

```python
row = int(input("Number of rows: "))

for i in range(1,row+1):
    for j in range(1,i+1):
        print("*", end=" ")
        
    print("")
```

# Loop Control Statement
break,continue,pass

```python
for i in range(1,10):
    if i ==5:
        break
    print(i)
```

```python
for i in range(1,10):
    if i ==5:
        continue
    print(i)
```

```python
# pass

for i in range(1,10):
    pass
```
